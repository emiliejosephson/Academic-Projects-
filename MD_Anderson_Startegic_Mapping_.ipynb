```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Capital Plan', 'Facilities', 'Research', 'Patient Care', 'Institutional Support', 'Instruction']
current_investment = [2440785478, 603017139, 943836403, 3764483888, 212830748, 330221014]
recommended_change = [0.2, -0.2, -0.1, -0.05, 0, 0]
complexity = [3, 2, 4, 5, 1, 2]  # Mock complexity values for each category

recommended_investment = [current * (1 + change) for current, change in zip(current_investment, recommended_change)]

# Normalize data for plotting
max_investment = max(current_investment + recommended_investment)
current_investment_norm = [x / max_investment for x in current_investment]
recommended_investment_norm = [x / max_investment for x in recommended_investment]
max_complexity = max(complexity)
complexity_norm = [c / max_complexity for c in complexity]
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_aspect('equal')

## Create circles with positions based on cost (x) and complexity (y)
for i, category in enumerate(categories):
    x_pos = current_investment_norm[i]  # Position on the x-axis (cost)
    y_pos = complexity_norm[i]          # Position on the y-axis (complexity)

    circle_current = plt.Circle((x_pos, y_pos), current_investment_norm[i] / 2, color='blue', alpha=0.5, label='Current' if i == 0 else "")
    circle_recommended = plt.Circle((x_pos, y_pos), recommended_investment_norm[i] / 2, color='green', alpha=0.5, label='Recommended' if i == 0 else "")

    ax.add_artist(circle_current)
    ax.add_artist(circle_recommended)
    ax.text(x_pos, y_pos, category, ha='center', va='center')


plt.legend(loc='upper right')

ax.set_xlabel('Costs')
ax.set_ylabel('Perceived Complexity')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)


ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)
plt.title('Strategic Group Mapping: Investment vs Complexity')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

years = [2018, 2019, 2020, 2021]

data = {
    'Year': years,
    'MD Anderson Admissions': [29118, 30399, 25748, 27082],
    'MSK Admissions': [24243, 25597, 22822, 24142],
    'MD Anderson Revenue': [4084154519, 4573840287, 4347374204, 4712140838],
    'MSK Revenue': [3539760000, 3973778000, 4560174000, 5011551000],
    'MD Anderson Length of Stay': [7.1, 7.2, 7.6, 7.5],
    'MSK Length of Stay': [7.1, 6.8, 7.1, 7.1]
}

df = pd.DataFrame(data)

fig, axes = plt.subplots(3, 1, figsize=(12, 12))

axes[0].plot(df['Year'], df['MD Anderson Admissions'], label='MD Anderson Admissions', marker='o', color='blue')
axes[0].plot(df['Year'], df['MSK Admissions'], label='MSK Admissions', marker='o', color='orange')
axes[0].set_title('Admissions Over Time', fontsize=14)
axes[0].set_ylabel('Admissions')
axes[0].legend()

# Revenue line plot
axes[1].plot(df['Year'], df['MD Anderson Revenue'], label='MD Anderson Revenue', marker='s', color='blue')
axes[1].plot(df['Year'], df['MSK Revenue'], label='MSK Revenue', marker='s', color='orange')
axes[1].set_title('Revenue Over Time', fontsize=14)
axes[1].set_ylabel('Revenue (in billions)')
axes[1].legend()


length_of_stay_data = np.array([df['MD Anderson Length of Stay'], df['MSK Length of Stay']])
sns.heatmap(length_of_stay_data, annot=True, fmt=".1f", cmap="coolwarm", ax=axes[2], xticklabels=years,
            yticklabels=['MD Anderson Length of Stay', 'MSK Length of Stay'])
axes[2].set_title('Average Length of Stay ', fontsize=14)


plt.tight_layout()
plt.show()
```

```python
##  Radar Chart

import numpy as np
import matplotlib.pyplot as plt
from math import pi


labels = ['Revenue', 'Admissions', 'Length of Stay']
years = [2018, 2019, 2020, 2021]


md_anderson_data = {
    'Revenue': [4084154519, 4573840287, 4347374204, 4712140838],
    'Admissions': [29118, 30399, 25748, 27082],
    'Length of Stay': [7.1, 7.2, 7.6, 7.5]
}

msk_data = {
    'Revenue': [3539760000, 3973778000, 4560174000, 5011551000],
    'Admissions': [24243, 25597, 22822, 24142],
    'Length of Stay': [7.1, 6.8, 7.1, 7.1]
}

# Function to normalize the values (0 to 1 scale for plotting)
def normalize(data):
    return [(float(i) - min(data)) / (max(data) - min(data)) for i in data]

## Normalized data for both institutions
md_anderson_normalized = {
    key: normalize(value) for key, value in md_anderson_data.items()
}

msk_normalized = {
    key: normalize(value) for key, value in msk_data.items()
}

md_anderson_values = np.array([md_anderson_normalized['Revenue'],
                               md_anderson_normalized['Admissions'],
                               md_anderson_normalized['Length of Stay']]).T

msk_values = np.array([msk_normalized['Revenue'],
                       msk_normalized['Admissions'],
                       msk_normalized['Length of Stay']]).T

num_vars = len(labels)


#Plotting

angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]


fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot for the years (MD Anderson)
for i in range(len(years)):
    ax.plot(angles, np.append(md_anderson_values[i], md_anderson_values[i][0]),
            label=f'MD Anderson {years[i]}', linewidth=2)

# Plot for the years (MSK)
for i in range(len(years)):
    ax.plot(angles, np.append(msk_values[i], msk_values[i][0]),
            label=f'MSK {years[i]}', linewidth=2, linestyle='dashed')

# Fill
for i in range(len(years)):
    ax.fill(angles, np.append(md_anderson_values[i], md_anderson_values[i][0]),
            alpha=0.1)

for i in range(len(years)):
    ax.fill(angles, np.append(msk_values[i], msk_values[i][0]), alpha=0.1)


ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Add a title
ax.set_title("MD Anderson vs MSK: Revenue, Admissions, and Length of Stay", size=15, color='navy', y=1.1)


plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```
